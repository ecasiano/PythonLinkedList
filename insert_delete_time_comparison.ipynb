```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Store worldline sizes
test_sizes = np.array([5,50,500,5000,50000,500000,5000000])
test_sizes

# Store elased times for 500000 insertions/deletions
elapsed_times_dll =  [3.2160,3.1402,3.2917,3.6037,4.0566,4.4012,5.0697]
elapsed_times_list = [1.1809,1.1156,1.2653,1.9973,10.5330,103.7240,]

# Store creation times for worldlines of various sizes
creation_times_dll =  [0.0715,0.0715,0.2150,0.3707,0.7868,5.4062,49.9266]
creation_times_list = [0.0000,0.0003,0.0028,0.0251,0.6643,32.0485,]
```

```python
# Plot
plt.plot(test_sizes[:-1],elapsed_times_list,label='Python List')
plt.plot(test_sizes,elapsed_times_dll,label='Doubly Linked List')
plt.xlabel('initial worldline size')
plt.ylabel('elapsed time (s)')
plt.xscale('log')
#plt.yscale('log')
plt.legend()
plt.show()
```

```python
plt.plot(test_sizes[:-1],creation_times_list,label='Python List')
plt.plot(test_sizes,creation_times_dll,label='Doubly Linked List')
plt.xlabel('initial worldline size')
plt.ylabel('creation time (s)')
plt.xscale('log')
plt.yscale('log')
plt.legend()
plt.show()
```
